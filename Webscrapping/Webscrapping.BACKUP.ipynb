###Ejemplos de datos en la columna 'units'


```
[{'price': '$3,664+', 'beds': '0'}, {'price': '$4,000+', 'beds': '1'}, {'price': '$5,630+', 'beds': '2'}, {'price': '$7,108+', 'beds': '3'}]
```
Este arreglo de diccionarios o Jsons representa que en la misma direccion existen 4 propiedades diferentes en renta. 

Para lo cual podemos hacer varias sustituciones con para tener separadores especiales. Uno para separa en renglones y otro en columnas

Buscamos algo parecido :
```
[{'price': '$3,664+', 'beds': '0'}
 {'price': '$4,000+', 'beds': '1'}
  {'price': '$5,630+', 'beds': '2'}
   {'price': '$7,108+', 'beds': '3'}]
```
<p>Vamos a todos los siguientes caracteres(innecesarios):"[{'+}]"

<p>Iniciemos separarando los datos con el comando explode de Pandas. 

<p>En este caso para convertir una columna necesitamos identificar un separador. 
<p>Por lo cual empleamos el comando split primero para indicar una cadena separadora: '}, {' 
<p>Posteriormente llamamos a explode para separar en renglones los datos.
<p>Se observaŕa el indice se repite, junto con todos los datos en las columnas restantes del registro.

````python
#Dejamos el DataFrame sin columnas y pasamos a REv1 (R1)
df_R1 = df_dropped

#Indiquemos el caracter o patron separador de los datos en la columna 'units'
df_R1['units'] = df_R1['units'].str.split('}, {')

#Ejecuta explode de PAndas para separar las listas de la columna en renglores
df_R1 =df_R1.explode('units').reset_index(drop=True)

````

Observamos que pasamos de 2000 registros a 3675

Al inicio observamos que este Dataset tenia 3 columna de IDs:
